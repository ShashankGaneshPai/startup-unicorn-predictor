```python
import pandas as pd
import numpy as np

df = pd.read_csv("../data/startup_valuation_dataset.csv")
```

```python
df["unicorn"] = (
    df["estimated_valuation_usd"] >= 1_000_000_000
).astype(int)
```

```python
df["unicorn"].value_counts()
```

```python
df.drop(
    columns=["estimated_valuation_usd"],
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
df.columns.tolist()
```

```python
for col in df.select_dtypes(include="object"):
    print(col, df[col].nunique())
```

```python
df.isnull().sum()

df.columns.tolist()

for col in df.select_dtypes(include="object"):
    print(col, df[col].nunique())
```

```python
df.drop(
    columns=[
        "startup_id",
        "startup_name"
    ],
    inplace=True
)
```

```python
df["funding_date"].head()
```

```python
df["funding_date"] = pd.to_datetime(
    df["funding_date"]
)
```

```python
df["funding_year"] = df["funding_date"].dt.year

df["funding_month"] = df["funding_date"].dt.month
```

```python
df.drop(
    columns=["funding_date"],
    inplace=True
)
```

```python
CURRENT_YEAR = 2026

df["startup_age"] = (
    CURRENT_YEAR -
    df["founded_year"]
)
```

```python
df["investor_count"] = (
    df["co_investors"]
    .fillna("")
    .apply(
        lambda x: len(str(x).split(","))
    )
)
```

```python
df.drop(
    columns=["co_investors"],
    inplace=True
)
```

```python
df.drop(
    columns=["tags"],
    inplace=True
)
```

```python
df.isnull().sum()
```

```python
df["exit_type"].value_counts(dropna=False)
```

```python
df["exit_type"] = df["exit_type"].fillna("No Exit")
```

```python
df["exit_type"].value_counts()
```

```python
df.drop(
    columns=[
        "exited",
        "exit_type"
    ],
    inplace=True
)
```

```python
num_cols = [
    "funding_amount_usd",
    "employee_count",
    "estimated_revenue_usd"
]

for col in num_cols:
    print(col)
    print(df[col].skew())
    print()
```

```python
import numpy as np

df["funding_amount_usd"] = np.log1p(
    df["funding_amount_usd"]
)

df["estimated_revenue_usd"] = np.log1p(
    df["estimated_revenue_usd"]
)
```

```python
for col in [
    "funding_amount_usd",
    "estimated_revenue_usd"
]:
    print(col)
    print(df[col].skew())
    print()
```

```python
cat_cols = df.select_dtypes(include="object").columns.tolist()

print(cat_cols)
```

```python
for col in cat_cols:
    print(f"\n{col}")
    print(df[col].nunique())
```

```python
country_freq = df["country"].value_counts(normalize=True)

df["country_freq"] = df["country"].map(
    country_freq
)
```

```python
df[["country","country_freq"]].head()
```

```python
df.drop(
    columns=["country"],
    inplace=True
)
```

```python
df = pd.get_dummies(
    df,
    columns=[
        "region",
        "industry",
        "funding_round",
        "lead_investor"
    ],
    drop_first=True
)
```

```python
df.shape
```

```python
df.head()
```

```python
df.to_csv(
    "processed_startup_data.csv",
    index=False
)

print("Processed dataset saved.")
```
